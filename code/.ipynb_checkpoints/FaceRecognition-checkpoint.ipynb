**2. PCA definition**

```python
class xu_PCA:
    
    def __init__(self):
        #Pixel point generally is used a variable range 0 to 255 to represent, so normalize data by divide 255 (range 0 to 1)
        self.std_num = 255.0
    
    def PCA(self, raw_data, PCs):
        data = raw_data/self.std_num
        S = np.cov(data.T)
        eigvalues,eigvectors = np.linalg.eig(S)
        indexByValues = np.argsort(eigvalues)[::-1] # sorted by eigenvalues biger to smaller
    
        eigvalues,eigvectors = eigvalues[indexByValues],eigvectors[:,indexByValues]
        eigvalues=np.real(eigvalues)
        eigvectors=np.real(eigvectors)
        subspace = data @ eigvectors[:,:PCs]
        return eigvalues,eigvectors,subspace

    #values,vectors,subs=PCA(X_std,30)
    #print(subs)

    def Plot_img(self, data):  # for this project assignment(Image recognition) only
        try:
            plt.imshow(data)
        except TypeError:
            print("check data structure, ** x **")

    #Plot_img(X_hat[0,:])

    def Reconstruct(self, data, eig_vectors,PCs):
        eig_vectors = eig_vectors[:,:PCs]
        cover_data = data @ eig_vectors.T
        cover_data = self.std_num * cover_data

        return cover_data

    # temp = Reconstruct(Z,eig_vectors[:,:30])
    # temp = temp[0,:].reshape((28,28))
    # Plot_img(temp)
    
    def MSE(self, data, raw_data):
        return np.mean(pow((raw_data-data),2))
    
    def PCA_noStd(self, data, PCs): 
        S = np.cov(data.T)
        eigvalues,eigvectors = np.linalg.eig(S)
        indexByValues = np.argsort(eigvalues)[::-1]

        eigvalues,eigvectors = eigvalues[indexByValues],eigvectors[:,indexByValues]
        eigvalues=np.real(eigvalues)
        eigvectors=np.real(eigvectors)
        subspace = data @ eigvectors[:,:PCs]
        return eigvalues,eigvectors,subspace
```

```python
# file loading func
def readbyLines_multi(filepath,header='no'):

    inputfile = open(filepath)
    data = []
    for line in inputfile:
        A = line.strip("\t").split()
        data.append(A)
    # remove header/column name
    if header == 'yes':
        data = data[1:]

    size = len(data)   
    inputfile.close()
    return data,size
```

```python
def accuracy(label,pred_result):
    pred_result = np.argmax(pred_result,axis=1)
    if len(label) != len(pred_result):
        return False
    count = 0
    for i in range(len(label)):
        if label[i] == pred_result[i]:
            count+=1 
    return count/len(label)
```

+ Convolution layer description: pytorch interface:Conv2d [5]

![conv2d](Conv2d_formula.jpg)

In the above formula, weight(C_outj,k) represents kerner j , k represents in_channel k, N_i represents batch i, so the part of sigma/sum is kerner j do operation of '\*\'  with input batch i data in k in_channel.  

Intuitively, this layer shared the same bias in the same out_channels(kernel)  

>For example: in_channels(k): RGB-> in_channel=3 ; Gray-> in_channel=1   
out_channels(Cout): in_channels * n_Kernels/Filters/Weights   

Other params of Conv2d:  
In this case, we set dilation default 1. There's the detail about dilation[9];  
And we only use 'groups = in_channels',that means n_filter = out_channels/in_channels


  **A Comprehensive Guide to Convolutional Neural Networks  the ELI5 way[8]** is very specified explainating that!


**1. Example of LeNet5, step by step forward building an network.**  
We used Relu as activation function in conv layers and the strategy of pooling is maximum.

```python
# Convolution layer1 - C1
conv_layer1 = nn.Conv2d(in_channels=1,out_channels=16,kernel_size=5, stride=2,padding = 1,padding_mode ='zeros',groups= 1)
in_data =Variable(torch.FloatTensor(ORL_train[0])).reshape(64,64)
print('Toolbox input structure :batch_size, in_channel, in_hight(rows), in_width(cols)\n',in_data.unsqueeze(0).unsqueeze(0).shape)
conv_output1 = F.relu(conv_layer1(in_data.unsqueeze(0).unsqueeze(0)))
print('Toolbox output structure :batch_size, out_channel, out_hight(rows), out_width(cols)\n',conv_output1.shape)
```

> We used one padding of zeros mode and the size of kernel is 5 and 16 different kerners, then getting 16 different feature maps with the shape of (31,31)

```python
# Max pooling layer1 - S2 
max_layer1 = nn.MaxPool2d(3, stride=2)
max_output1 = max_layer1(conv_output1)
print(max_output1.shape)
```

> We used 3\*3 window size to max from feature maps with stride 2, then getting 15\*15 pooling maps 

```python
# Convolution layer2 - C3
conv_layer2 = nn.Conv2d(in_channels=16,out_channels=32,kernel_size=5, stride=1,padding = 1,padding_mode ='zeros',groups= 16)
conv_output2 = F.relu(conv_layer2(max_output1))
print(conv_output2.shape)
```

> 16 pooled maps to be convoluted with 2 different filters and then generated 32 feature maps with the shape of (13,13)

```python
# Max pooling layer2 - S4
max_layer2 = nn.MaxPool2d(2, stride=2)
max_output2 = max_layer2(conv_output2)
print(max_output2.shape)
```

> We used 2\*2 window size to max from feature maps with stride 2, then getting 6\*6 pooling maps 

```python
# Fully-connected layer/Convolution layer - C5 

batch_size=1  # we just put one image for debug 
fc_in = max_output2.view(batch_size, -1) #  flatten
fc1 = nn.Linear(32*6*6, 120) # given 100 filers with the same size(single map) of S4 output
fc1_output = F.relu(fc1(fc_in))
```

> Layers S4 and C5 are fully-connected.That is also why in some implementations of LeNet-5 actually use a fully-connected layer instead of the convolutional one as the 5th layer[7].  C5 is labeled as a convolutional layer instead of a fully connected layer, because if LeNet-5 input becomes larger and its structure remains unchanged, its output size will be greater than 1x1, i.e. not a fully connected layer[10].

```python
# Fully-connected layer - F6
fc2 = nn.Linear(120, 80) # 120 node to 80 node

fc2_output = F.relu(fc2(fc1_output))
print(fc2_output.shape)
```

```python
# Fully-connected layer - F7
n_labels_ORL = 40 
dropout_layer = nn.Dropout(0.1) # enhance generalization ability
fc3 = nn.Linear(80, n_labels_ORL) # 80 node to 40 classes 
fc3_output = fc3(dropout_layer(fc2_output))
print(fc3_output.shape)
```

```python
# softmax output layer
softmax_layer = nn.Softmax(dim=0)
result = softmax_layer(fc3_output.squeeze(0))
print(result)
print('the prediction class is:',result.argmax())
```

**2.Building model of LeNet5 with train process**  

```python
class CNN(nn.Module):
    def __init__(self,cat_labels,dropout_rate):
        super(CNN,self).__init__()
        
        self.conv_layer1 = nn.Conv2d(in_channels=1,out_channels=16,kernel_size=5, stride=2,padding = 1,padding_mode ='zeros',groups= 1)
        self.max_layer1 = nn.MaxPool2d(3, stride=2)
        self.conv_layer2 = nn.Conv2d(in_channels=16,out_channels=32,kernel_size=5, stride=1,padding = 1,padding_mode ='zeros',groups= 16)
        self.max_layer2 = nn.MaxPool2d(2, stride=2)

        self.fc1 = nn.Linear(32*6*6, 120)
        self.fc2 = nn.Linear(120, 80)
        self.fc3 = nn.Linear(80, cat_labels)
        
        self.dropout = nn.Dropout(dropout_rate)  
        self.softmax = nn.Softmax(dim=1)
        self.relu = nn.ReLU()
        
    def forward(self,data):
        C1 = self.relu(self.conv_layer1(data))
        S2 = self.max_layer1(C1)
        C3 = self.relu(self.conv_layer2(S2))
        S4 = self.max_layer2(C3)
        F_in   = S4.view(data.size(0),-1) # flatten
        F5 = self.relu(self.fc1(F_in))
        F6 = self.relu(self.fc2(F5))
        F7 = self.fc3(self.dropout(F6))
        result = self.softmax(F7)
        return result
```

```python
# model instance / optimizer intialization
cnn_model = CNN(40,0.1)
#optimizer = torch.optim.Adadelta(cnn_model.parameters(), rho = 0.95, weight_decay = 0.001)
optimizer = torch.optim.SGD(cnn_model.parameters(), lr=0.01, momentum=0.9)
```

```python
# training data load
ORL64_train=Variable(torch.FloatTensor(ORL_train)).reshape(80,64,64).unsqueeze(1)
ORL64_label=Variable(torch.LongTensor(ORL_train_y-1).squeeze(1))

# some parameters
n_epoch=3000
best_epoch = 0
best_loss = 1111111
out_dir = r"\outfiles"
loss_epoch = []

# train model
for epoch in range(n_epoch):
    optimizer.zero_grad()
    model_out = cnn_model(ORL64_train)
    loss = F.cross_entropy(model_out, ORL64_label)
    loss.backward()
    optimizer.step()
    loss_epoch.append(loss.data)
    
    if loss.data < best_loss:
        best_loss = loss.data
        best_epoch = epoch
        torch.save(cnn_model.state_dict(), main_dir + out_dir + '\ORL64_cnn.pth' )
```

> Relevant function

```python
class CNN_32(nn.Module):
    def __init__(self,cat_labels,dropout_rate):
        super(CNN_32,self).__init__()
        
        self.conv_layer1 = nn.Conv2d(in_channels=1,out_channels=8,kernel_size=5, stride=1,padding = 1,padding_mode ='zeros',groups= 1)
        self.max_layer1 = nn.MaxPool2d(3, stride=2)
        self.conv_layer2 = nn.Conv2d(in_channels=8,out_channels=16,kernel_size=5, stride=1,padding = 1,padding_mode ='zeros',groups= 8)
        self.max_layer2 = nn.MaxPool2d(2, stride=2)

        self.fc1 = nn.Linear(16*6*6, 120)  # 32*6*6 is from maxpooling layer 2
        self.fc2 = nn.Linear(120, 80)
        self.fc3 = nn.Linear(80, cat_labels)
        
        self.dropout = nn.Dropout(dropout_rate)  
        self.softmax = nn.Softmax(dim=1)
        self.relu = nn.ReLU()
        
    def forward(self,data):
        C1 = self.relu(self.conv_layer1(data))
        S2 = self.max_layer1(C1)
        C3 = self.relu(self.conv_layer2(S2))
        S4 = self.max_layer2(C3)
        F_in   = S4.view(data.size(0),-1) # flatten
        F5 = self.relu(self.fc1(F_in))
        F6 = self.relu(self.fc2(F5))
        F7 = self.fc3(self.dropout(F6))
        result = self.softmax(F7)
        return result

class CNN_64(nn.Module):
    def __init__(self,cat_labels,dropout_rate):
        super(CNN_64,self).__init__()
        
        self.conv_layer1 = nn.Conv2d(in_channels=1,out_channels=16,kernel_size=5, stride=2,padding = 1,padding_mode ='zeros',groups= 1)
        self.max_layer1 = nn.MaxPool2d(3, stride=2)
        self.conv_layer2 = nn.Conv2d(in_channels=16,out_channels=32,kernel_size=5, stride=1,padding = 1,padding_mode ='zeros',groups= 16)
        self.max_layer2 = nn.MaxPool2d(2, stride=2)

        self.fc1 = nn.Linear(32*6*6, 120) 
        self.fc2 = nn.Linear(120, 80)
        self.fc3 = nn.Linear(80, cat_labels)
        
        self.dropout = nn.Dropout(dropout_rate)  
        self.softmax = nn.Softmax(dim=1)
        self.relu = nn.ReLU()
        
    def forward(self,data):
        C1 = self.relu(self.conv_layer1(data))
        S2 = self.max_layer1(C1)
        C3 = self.relu(self.conv_layer2(S2))
        S4 = self.max_layer2(C3)
        F_in   = S4.view(data.size(0),-1) # flatten
        F5 = self.relu(self.fc1(F_in))
        F6 = self.relu(self.fc2(F5))
        F7 = self.fc3(self.dropout(F6))
        result = self.softmax(F7)
        return result
    
def test_accuary(param_path,c_labels,testdata,testlabel,index_dir,db,sets,mode='64'):
    reload_states = torch.load(param_path)

    # model load
    if mode == '64':
        test_model = CNN_64(c_labels,0.1)
        n_pix = 64
    elif mode == '32':
        test_model = CNN_32(c_labels,0.1)
        n_pix = 32
    test_model.load_state_dict(reload_states)

    # testing data load
    splited_ind = loadmat(main_dir+data_dir+'\\'+ db+ '\\'+sets+'\\1.mat')
    ind_train = splited_ind['trainIdx'].squeeze()
    ind_test = splited_ind['testIdx'].squeeze()
    train,train_y,test,test_y = SetsSplit(testdata,testlabel,ind_train,ind_test)
    test=Variable(torch.FloatTensor(test)).reshape(len(test),n_pix,n_pix).unsqueeze(1)
    label=Variable(torch.LongTensor(test_y-1).squeeze(1))

    # classification
    pred_test = test_model(test).argmax(dim=1)

    # accuracy calculation
    count=0
    for i in range(len(pred_test)):
        if pred_test[i]== label[i]:
            count+=1
    acc = count/len(pred_test)
    return (acc)
```
